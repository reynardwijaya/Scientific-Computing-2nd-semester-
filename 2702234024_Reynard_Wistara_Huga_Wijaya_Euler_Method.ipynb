```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
%matplotlib inline


# Define parametersl
f = lambda t, s: np.exp(-t) # ODE
h = 0.1 # Step size
t = np.arange(0, 1 + h, h) # Numerical grid
s0 = -1 # Initial Condition

# Explicit Euler Method
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
  s[i + 1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (12, 8))
plt.plot(t, s, 'bo--', label='Approximate')
plt.plot(t, -np.exp(-t), 'g', label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

# y' + 4y  = x^2, y(0) = 1

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function for the differential equation y' + 4y = x^2
def f(x, y):
    return x**2 - 4*y

# Parameters
h = 0.01  # Step size
t = np.arange(0, 1 + h, h)  # Numerical grid
y0 = 1  # Initial Condition

# Explicit Euler Method
y = np.zeros(len(t))
y[0] = y0

for i in range(0, len(t) - 1):
    y[i + 1] = y[i] + h * f(t[i], y[i])

# Exact Solution
exact_y = (1/16) * (1 - np.exp(-4*t)) + (t*3) / 12 - (t*2) / 8 + t / 32

# Plotting
plt.figure(figsize=(12, 8))
plt.plot(t, y, 'b--', label='Approximate')
plt.plot(t, exact_y, 'g', label='Exact')
plt.title("Approximate and Exact Solution \n for the ODE $y' + 4y = x^2$")
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-poster')  # Use the updated style name
%matplotlib inline

# Define parameters
f = lambda t, s: t**2 - 4 * s  # ODE
h = 0.1  # Step size
t = np.arange(0, 1 + h, h)  # Numerical grid
s0 = 1  # Initial Condition

# Explicit Euler Method
s = np.zeros(len(t))
s[0] = s0

for i in range(len(t) - 1):
    s[i + 1] = s[i] + h * f(t[i], s[i])

# Define the exact solution function
exact_solution = lambda t: (31 / 32) * np.exp(-4 * t) + (1 / 4) * t**2 - (1 / 8) * t + (1 / 32)

# Plotting
plt.figure(figsize=(12, 8))
plt.plot(t, s, 'bo--', label='Approximate')
plt.plot(t, exact_solution(t), 'g', label='Exact')
plt.title('Approximate and Exact Solution for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

# y' = sin y,      y(0) = 1

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
%matplotlib inline

# Define parameters and functions
f = lambda y: np.sin(y)  # ODE
h = 0.1  # Step size
x = np.arange(0, 0.51, h)  # Numerical grid
y0 = 1  # Initial Condition

# Explicit Euler Method
y = np.zeros(len(x))
y[0] = y0

for i in range(len(x) - 1):
    y[i + 1] = y[i] + h * f(y[i])

# Plotting
plt.figure(figsize=(12, 8))
plt.plot(x, y, 'bo--', label='Approximate')
plt.title('Approximate Solution for y\' = sin(y) with y(0) = 1')
plt.xlabel('x')
plt.ylabel('y(x)')
plt.grid()
plt.legend(loc='upper left')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-poster') # Use the updated style name
%matplotlib inline

# Define parameters
f = lambda t, s: np.exp(-t) # ODE
h = 0.1 # Step size
t = np.arange(0, 1 + h, h) # Numerical grid
s0 = -1 # Initial Condition

# Explicit Euler Method
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h * f(t[i], s[i])

# Define the inverse function
x = lambda y: np.log(1/np.sin(y) - np.cos(y)/np.sin(y)) + 0.604582

# Plotting
plt.figure(figsize=(12, 8))
plt.plot(s, t, 'bo--', label='Approximate')  # Note: swapping x and y for plotting
plt.plot(-np.exp(-t), t, 'g', label='Exact')  # Note: swapping x and y for plotting
plt.title('Approximate and Exact Solution for Simple ODE')
plt.xlabel('f(t)')
plt.ylabel('t')  # Note: swapping x and y for plotting
plt.grid()
plt.legend(loc='lower right')
plt.show()

# Plot the inverse function
y_values = np.linspace(-np.pi/2 + 0.1, np.pi/2 - 0.1, 100)  # Avoiding singularities
x_values = x(y_values)
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, 'r-', label='Inverse Function')
plt.title('Inverse Function x(y)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()
```
